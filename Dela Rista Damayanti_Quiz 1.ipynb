1. Object Oriented Programming adalah sebuah paradigma pemrograman yang merepresentasikan objek seperti objek yang ada di sekitar kehidupan sehari-hari ke dalam bentuk kode program. Contoh implementasi dibawah

```python
#Encapsulation
class Novel(object):
    def __init__(self, judul, pengarang, penerbit, tahun_terbit):
        self.judul = judul
        self.pengarang = pengarang
        self.penerbit = penerbit
        self.tahun_terbit = tahun_terbit
    def pinjam(self):
        print("Pinjam maksimal 3 hari, judul novel", self.judul)
    def Info(self):
        print(f"Judul: {self.judul}, Pengarang: {self.pengarang}, Penerbit: {self.penerbit}, Tahun Terbit: {self.tahun_terbit}")
        
novel1 = Novel("Buku Minta Di Sayang", "Rintik Sedu", "Gramedia", "2019")

novel1.Info()
novel1.pinjam()
    
```

```python
#Inheritence

class Book(object):
    def __init__(self, judul, pengarang, penerbit, tahun_terbit):
        self.judul = judul
        self.pengarang = pengarang
        self.penerbit = penerbit
        self.tahun_terbit = tahun_terbit
    def durasi_pinjam(self, durasi):
        print("Lama durasi pinjam: ", durasi)
    
    def info_book(self):
        print(f"Judul: {self.judul}, Pengarang: {self.pengarang}, Penerbit: {self.penerbit}, Tahun Terbit: {self.tahun_terbit}")
        self.durasi_pinjam(3)
        
buku1 = Book("Kata", "Rintik Sedu", "Gramedia", "2017")

buku1.info_book()
```

```python
class Majalah(Book):
    def __init__(self, judul, pengarang, penerbit, tahun_terbit, edisi):
        super().__init__(judul, pengarang, penerbit, tahun_terbit)
        self.edisi = edisi
    
    def info_majalah(self):
        super().info_book()
        print(f"edisi: {self.edisi}")

majalah1 = Majalah("Majalah Bobo", "Stevani", "Baby Boy", "2008", "9")

majalah1.info_majalah()
```
